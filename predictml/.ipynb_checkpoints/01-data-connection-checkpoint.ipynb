```python
import yfinance as yf
import pandas as pd
import os
from datetime import datetime, timedelta

# Compressed file path
filename = "market.csv.gz"

# Check if file exists and load existing data
if os.path.exists(filename):
    print(f"Found existing file: {filename}")
    existing_data = pd.read_csv(filename, index_col=0, parse_dates=True)
    last_date = existing_data.index[-1]
    print(f"Last date in existing data: {last_date}")
    
    # Fetch only new data from the day after last_date
    ticker = yf.Ticker("^SPX")
    start_date = (last_date + timedelta(days=1)).strftime('%Y-%m-%d')
    new_data = ticker.history(start=start_date)
    
    if len(new_data) > 0:
        print(f"\nFetched {len(new_data)} new records")
        print(f"New data range: {new_data.index[0]} to {new_data.index[-1]}")
        
        # Append new data to existing data
        data = pd.concat([existing_data, new_data])
        print(f"\nUpdated data shape: {data.shape}")
        print(f"Full date range: {data.index[0]} to {data.index[-1]}")
    else:
        print("\nNo new data available. Using existing data.")
        data = existing_data
else:
    print(f"No existing file found. Fetching full history...")
    # Fetch historical data for maximum available period
    ticker = yf.Ticker("^SPX")
    data = ticker.history(period="max")
    
    print(data.head())
    print(f"\nData shape: {data.shape}")
    print(f"Date range: {data.index[0]} to {data.index[-1]}")

# Save as compressed CSV
data.to_csv(filename, compression='gzip')
print(f"\nData saved to {filename} (compressed)")

# Plot the closing prices
data['Close'].plot(title="Historical Closing Prices", ylabel="Price", xlabel="Date")
```
