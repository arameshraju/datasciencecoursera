```python
import yfinance as yf

# Fetch historical data for maximum available period
ticker = yf.Ticker("^SPX")  # Replace "AAPL" with your desired ticker symbol
data = ticker.history(period="max")

# Display the data
print(data.head())
print(f"\nData shape: {data.shape}")
print(f"Date range: {data.index[0]} to {data.index[-1]}")

data['Close'].plot(title="Historical Closing Prices", ylabel="Price", xlabel="Date")
```
