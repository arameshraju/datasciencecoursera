```python
import pandas as pd
import numpy as np

# load spc_astro_data.csv
spx_dates_df = pd.read_csv('spx_astro_data.csv')
# make data as index 
spx_dates_df.set_index('Date', inplace=True)
# if value  > 0 then 1 else 0
spx_dates_df['change_pc']= np.where(spx_dates_df['change_pc']>0,1,0)
spx_dates_df.head()
```

```python
import pandas as pd

dfn= spx_dates_df.copy()
SIGNS = [
    'Aries', 'Taurus', 'Gemini', 'Cancer', 'Leo', 'Virgo',
    'Libra', 'Scorpio', 'Sagittarius', 'Capricorn', 'Aquarius', 'Pisces'
]
# create a map for Moon_Sign to numerical values
dfn['Ascendant'] = dfn['Ascendant'].apply(lambda x: SIGNS.index(x))

dfn['Sun_Sign'] = dfn['Sun_Sign'].apply(lambda x: SIGNS.index(x))
dfn['Moon_Sign'] = dfn['Moon_Sign'].apply(lambda x: SIGNS.index(x))
dfn['Mars_Sign'] = dfn['Mars_Sign'].apply(lambda x: SIGNS.index(x))
dfn['Mercury_Sign'] = dfn['Mercury_Sign'].apply(lambda x: SIGNS.index(x))
dfn['Jupiter_Sign'] = dfn['Jupiter_Sign'].apply(lambda x: SIGNS.index(x))
dfn['Venus_Sign'] = dfn['Venus_Sign'].apply(lambda x: SIGNS.index(x))
dfn['Saturn_Sign'] = dfn['Saturn_Sign'].apply(lambda x: SIGNS.index(x))
dfn['Rahu_Sign'] = dfn['Rahu_Sign'].apply(lambda x: SIGNS.index(x))
dfn['Ketu_Sign'] = dfn['Ketu_Sign'].apply(lambda x: SIGNS.index(x))

dfn.head()
```

```python
df= spx_dates_df.copy()
df.head()
```

```python
# filter change_pc == 1 
import pandas as pd
cdff = df[df['change_pc'] == 1]
cdff.head()
# group by Ascendant and sum of change_pc
cordata = cdff.groupby('Ascendant')['change_pc'].sum()
# rename  Ascendant as sign_name
cordata = cordata.rename_axis('sign_name')
# rename series to ascendant
cordata = cordata.rename('ascendant')
cordata

```

```python
platnets = [
    'Moon_Sign', 'Sun_Sign', 'Mars_Sign', 'Mercury_Sign',
    'Jupiter_Sign', 'Venus_Sign', 'Saturn_Sign', 'Rahu_Sign', 'Ketu_Sign'
]

final_df =cordata.copy()

for planet in platnets:
    # group by planet and sum of change_pc
    pdata = cdff.groupby(planet)['change_pc'].sum()
    # rename planet as sign_name
    pdata = pdata.rename_axis('sign_name')
    # rename series to planet
    pdata = pdata.rename(planet)
    # join cordata and pdata on sign_name
    final_df = pd.concat([final_df, pdata], axis=1)
# sdata = cdff.groupby('Moon_Sign')['change_pc'].sum()
# sdata = sdata.rename_axis('sign_name')
# sdata = sdata.rename('Moon_Sign')

# join cordata and sdata on sign_name
final_df
```

```python
# plot sign_name vs all planets sum of change_pc
import seaborn as sns
import matplotlib.pyplot as plt
final_df.reset_index(inplace=True)
plt.figure(figsize=(12, 8))
sns.lineplot(data=final_df, x='sign_name', y='ascendant', marker='o', label='Ascendant')
sns.lineplot(data=final_df, x='sign_name', y='Moon_Sign', marker='o', label='Moon_Sign')
sns.lineplot(data=final_df, x='sign_name', y='Sun_Sign', marker='o', label='Sun_Sign')
sns.lineplot(data=final_df, x='sign_name', y='Mars_Sign',   marker='o', label='Mars_Sign')
sns.lineplot(data=final_df, x='sign_name', y='Mercury_Sign', marker='o', label='Mercury_Sign')
sns.lineplot(data=final_df, x='sign_name', y='Jupiter_Sign', marker='o', label='Jupiter_Sign')
sns.lineplot(data=final_df, x='sign_name', y='Venus_Sign', marker='o', label='Venus_Sign')
sns.lineplot(data=final_df, x='sign_name', y='Saturn_Sign', marker='o', label='Saturn_Sign')
sns.lineplot(data=final_df, x='sign_name', y='Rahu_Sign', marker='o', label='Rahu_Sign')
sns.lineplot(data=final_df, x='sign_name', y='Ketu_Sign', marker='o', label='Ketu_Sign')
plt.title('Sum of change_pc by Astrological Signs') 
```
