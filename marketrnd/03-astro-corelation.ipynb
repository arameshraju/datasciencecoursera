```python
import pandas as pd
import numpy as np

# load spc_astro_data.csv
spx_dates_df = pd.read_csv('spx_astro_data.csv')
# make data as index 
spx_dates_df.set_index('Date', inplace=True)
# if value  > 0 then 1 else 0
spx_dates_df['change_pc']= np.where(spx_dates_df['change_pc']>0,1,0)
spx_dates_df.head()
```

```python
import pandas as pd

df= spx_dates_df.copy()
# create a map for Moon_Sign to numerical values
SIGNS = [
    'Aries', 'Taurus', 'Gemini', 'Cancer', 'Leo', 'Virgo',
    'Libra', 'Scorpio', 'Sagittarius', 'Capricorn', 'Aquarius', 'Pisces'
]
df['Sun_Sign'] = df['Sun_Sign'].apply(lambda x: SIGNS.index(x))
df['Moon_Sign'] = df['Moon_Sign'].apply(lambda x: SIGNS.index(x))
df['Mars_Sign'] = df['Mars_Sign'].apply(lambda x: SIGNS.index(x))
df['Mercury_Sign'] = df['Mercury_Sign'].apply(lambda x: SIGNS.index(x))
df['Jupiter_Sign'] = df['Jupiter_Sign'].apply(lambda x: SIGNS.index(x))
df['Venus_Sign'] = df['Venus_Sign'].apply(lambda x: SIGNS.index(x))
df['Saturn_Sign'] = df['Saturn_Sign'].apply(lambda x: SIGNS.index(x))
df['Rahu_Sign'] = df['Rahu_Sign'].apply(lambda x: SIGNS.index(x))
df['Ketu_Sign'] = df['Ketu_Sign'].apply(lambda x: SIGNS.index(x))

df.head()
```

```python
# Calculate the correlation directly between the two columns
correlation_value = df['change_pc'].corr(df['Moon_Sign'])

print(f"Correlation between change_pc and Moon_Sign: {correlation_value}")
```
