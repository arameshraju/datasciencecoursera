```python
# connecto to yahoo finance and fetch the spx historical data for max period 
import yfinance as yf
import pandas as pd
import numpy as np
import datetime
import os

# Fetch SPX historical data for maximum period
spx = yf.Ticker("^SPX")
spx_hist = spx.history(period="max")
spx_hist.reset_index(inplace=True)
# add a new column change_pc to the dataframe to calculate the percentage change between the closing price of the current day and the previous day
```

```python
spx_hist['change_pc'] = spx_hist['Close'].pct_change() * 100
spx_hist.dropna(inplace=True)
spx_hist
```

```python
# save the dataframe to a csv file
spx_hist.to_csv("spx_historical_data.csv", index=False)
```
